```python
import pandas as pd
df = pd.read_excel("superstores_sales.csv")
df.head()
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
print(df.columns.tolist())
```

```python
df.columns = df.columns.str.strip()
print(df.columns.tolist())
```

```python
print("Sales" in df.columns)
print("profit" in df.columns)
```

```python
df.drop_duplicates(inplace=True)
df.dropna(subset=['Sales', 'Profit'], inplace=True)
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Category'] = df['Category'].str.strip()
df['Region'] = df['Region'].str.strip()
```

```python
df.info()
df.head()
```

```python
region_sales = df.groupby('Region')['Sales'].sum().reset_index()
region_profit = df.groupby('Region')['Profit'].sum().reset_index()

print(region_sales)
print(region_profit)
```

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
print(top_products)
```

```python
df['Month'] = df['Order Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Sales'].sum()
print(monthly_sales)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Region', y='Sales', data=df, estimator=sum)
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Category', y='Profit', data=df, estimator=sum)
plt.title('Total Profit by Category')
plt.xlabel('Category')
plt.ylabel('Total Profit')
plt.show()
```

```python
plt.figure(figsize=(8,4))
monthly_sales.plot(kind='line', marker='o', title='Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
top10_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(8,5))
top10_products.plot(kind='barh', color='skyblue')
plt.title('Top 10 Products by Sales')
plt.xlabel('Sales')
plt.ylabel('Product Name')
plt.gca().invert_yaxis()
plt.show()
```

```python
df['Profit Margin'] = (df['Profit'] / df['Sales']) * 100

plt.figure(figsize=(6,4))
sns.boxplot(x='Category', y='Profit Margin', data=df)
plt.title('Profit Margin by Category')
plt.ylabel('Profit Margin (%)')
plt.show()
```

```python
df.to_csv("superstores_sales_cleaned.csv",index = False)
```
